```python
import matplotlib.pyplot as plt
import numpy as np


# Fixing random state for reproducibility
np.random.seed(19680801)
i = 0
j = 0
k = 0
fig, axs = plt.subplots(2, 2)
cmaps = ['RdBu_r', 'viridis']
for col in range(2):
    for row in range(2):
        ax = axs[col, row]
        pcm = ax.pcolormesh(np.random.randint(i+k,j+k+2,(20, 20)),
                            cmap= cmaps[col])
        k+=2
        fig.colorbar(pcm, ax=ax)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import AxesGrid
import numpy as np


# Fixing random state for reproducibility
np.random.seed(19680801)
# _, _, height, width = t.shape
# figsize = width / float(dpi), height / float(dpi)
i = 0
j = 0
k = 0
fig = plt.figure()
grid = AxesGrid(fig, 111,
            nrows_ncols=(2, 2),
            label_mode=1,
            axes_pad=0.05,
            cbar_mode='single',
            cbar_location='right',
            cbar_pad=0.1,
            share_all=True
            )
count = 0
#looping through all the kernels in each channel
for ax in grid:
    pcm = ax.pcolormesh(np.random.randint(i+k,j+k+2,(20, 20)), vmin = 0, vmax = 8)
    ax.axis('off')
    k+=2
    count +=1
cbar = ax.cax.colorbar(pcm)
cbar = grid.cbar_axes[0].colorbar(pcm)
plt.show()
```
